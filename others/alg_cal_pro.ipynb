def mul_two_polynomials(p1,p2):
    temp_poly = []
    for i in range(0,len(p1)):
        for j in range(0,len(p2)):
            if p1[i][1]*p2[j][1] == 0:
                temp_poly.append([0,0])
            else:
                temp_poly.append([p1[i][0] + p2[j][0],p1[i][1]*p2[j][1]])
    temp_poly.sort()
    out_poly = [temp_poly[0]]
    for i in range(1,len(temp_poly)):
        if temp_poly[i][0] == out_poly[-1][0]:
            out_poly[-1][1] += temp_poly[i][1]
        else:
            out_poly.append(temp_poly[i])
    #print_polynomial(out_poly)
    return out_poly

def add_two_polynomials(p1,p2):
    temp_poly = [tp[:] for tp in p1]
    for i in range(0,len(p2)):
        #print_polynomial(temp_poly)
        has_same_degree = False
        for j in range(0,len(temp_poly)):
            if temp_poly[j][0] == p2[i][0]:
                temp_poly[j][1] = temp_poly[j][1] + p2[i][1]
                has_same_degree = True
                break
        if not has_same_degree:
            temp_poly.append(p2[i][:])
    #print_polynomial(temp_poly)
    out_poly = []
    for i in range(0,len(temp_poly)):
        if temp_poly[i][1] != 0:
            out_poly.append(temp_poly[i][:])
    if out_poly == []:
        out_poly.append([0,0])
    out_poly.sort()
    return out_poly
        

def get_mul_polynomial(k):
    mul_poly = []
    if k == 0:
        mul_poly.append([0,0])
    elif k > 0:
        ki = k - 1
        while(ki >= 1 - k):
            mul_poly.append([ki,1])
            ki -= 2
    else:
        ki = -k -1
        while(ki >= 1+k):
            mul_poly.append([ki,-1])
            ki -= 2
    #print_polynomial(mul_poly)
    return mul_poly
